## 0 - import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb

from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
from sklearn.model_selection import GridSearchCV, train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV, LogisticRegression, ElasticNet, ElasticNetCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def dataframe_info(df):
    # Initialiser une liste pour stocker les informations
    info_list = []

    # Parcourir chaque colonne pour recueillir les informations
    for col in df.columns:
        non_null_count = df[col].notnull().sum()
        dtype = df[col].dtype
        nan_count = df[col].isna().sum()
        nan_percentage = (nan_count / len(df)) * 100
        example_value = df[col].dropna().iloc[0] if non_null_count > 0 else None
        info_list.append([col, non_null_count, nan_count, round(nan_percentage, 2), dtype, example_value])

    # Créer un DataFrame à partir de la liste d'informations
    info_df = pd.DataFrame(info_list, columns=['Column', 'Non-Null Count', 'NaN Count', 'NaN Percentage', 'Dtype', 'Example Value'])
    
    return info_df
```

```python
df = pd.read_csv('./ML/clean_data.csv', low_memory = False)
```

```python
dataframe_info(df)
```

## 1 - Normalisation pre-split

```python
# Normalisation des variables numériques cycliques

df_norm = df

# Utilisez .loc pour faire des assignations
df_norm.loc[:, 'hour_sin'] = np.sin(2 * np.pi * df_norm['hour'] / 24)
df_norm.loc[:, 'hour_cos'] = np.cos(2 * np.pi * df_norm['hour'] / 24)

df_norm.loc[:, 'weekday_sin'] = np.sin(2 * np.pi * df_norm['weekday'] / 7)
df_norm.loc[:, 'weekday_cos'] = np.cos(2 * np.pi * df_norm['weekday'] / 7)

df_norm.loc[:, 'month_sin'] = np.sin(2 * np.pi * df_norm['month'] / 12)
df_norm.loc[:, 'month_cos'] = np.cos(2 * np.pi * df_norm['month'] / 12)

df_norm = df_norm.drop(columns = ['month','weekday','hour'])
```

## 2 - Encodage

```python
df_encoded = df_norm
```

```python
# Encodage binaire : DeployedFromLocation
encoder = LabelEncoder()
df_encoded['DeployedFromLocation'] = encoder.fit_transform(df_encoded['DeployedFromLocation'])
```

### Version sans DeployedFromStation_Name

```python
df_encoded = df_encoded.drop(columns = ["DeployedFromStation_Name"])
# Encodage getdummies : "PlusCode_Description", "PropertyCategory","AddressQualifier","IncidentType"
cols_to_encode = ["PlusCode_Description", "PropertyCategory","AddressQualifier","IncidentType"]
df_encoded = pd.get_dummies(df_encoded, columns=cols_to_encode)
```

### Version avec DeployedFromStation_Name

```python
# Encodage getdummies : "DeployedFromStation_Name","PlusCode_Description", "PropertyCategory","AddressQualifier","IncidentType"
cols_to_encode = ["DeployedFromStation_Name","PlusCode_Description", "PropertyCategory","AddressQualifier","IncidentType"]
df_encoded = pd.get_dummies(df_encoded, columns=cols_to_encode)
```

## 3 - Création des échantillons Train/Test

```python
target = df_encoded.ResponseDuration
X = df_encoded.drop(labels = ["ResponseDuration"], axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X,target, test_size=0.2, random_state=42)
```

## 4 - Normalisation (Standard & Minmax)

### Northing & Easting on Standard

```python
# Standard scaler à utiliser sur les variables à distribution normale "Northing_rounded","Easting_rounded"
columns_for_standard_scaling = ["Northing_rounded","Easting_rounded"]

scaler = StandardScaler()

X_train[columns_for_standard_scaling] = scaler.fit_transform(X_train[columns_for_standard_scaling])
X_test[columns_for_standard_scaling] = scaler.transform(X_test[columns_for_standard_scaling])
```

```python

# Afficher la distribution des variables après mise à l'échelle
plt.figure(figsize=(12, 6))
plt.suptitle('Distribution des variables après mise à l\'échelle (StandardScaler)', fontsize=16)

for i, col in enumerate(columns_for_standard_scaling, start=1):
    plt.subplot(2, 3, i)
    sns.histplot(X_train[col], bins=10, kde=True)
    plt.title(col)

plt.tight_layout()
plt.show()
```

### Northing and Easting on MinMax

```python
# Standard scaler à utiliser sur les variables à distribution normale "Northing_rounded","Easting_rounded"
columns_for_standard_scaling = ["Northing_rounded","Easting_rounded"]

scaler = MinMaxScaler()

X_train[columns_for_standard_scaling] = scaler.fit_transform(X_train[columns_for_standard_scaling])
X_test[columns_for_standard_scaling] = scaler.transform(X_test[columns_for_standard_scaling])
```

```python
# Afficher la distribution des variables après mise à l'échelle
plt.figure(figsize=(12, 6))
plt.suptitle('Distribution des variables après mise à l\'échelle (MinMaxScaler)', fontsize=16)

for i, col in enumerate(columns_for_standard_scaling, start=1):
    plt.subplot(2, 3, i)
    sns.histplot(X_train[col], bins=10, kde=True)
    plt.title(col)

plt.tight_layout()
plt.show()
```

### Reste de la standardisation

```python
# Min Max à utiliser sur des variables sns distribution particulières 
# "PumpOrder", "NumStationsWithPumpsAttending", "NumPumpsAttending", "NumCalls", "year"
columns_for_minmax_scaling = ["PumpOrder", "NumStationsWithPumpsAttending", "NumPumpsAttending", "NumCalls", "year"]

minmax = MinMaxScaler()

X_train[columns_for_minmax_scaling]= minmax.fit_transform(X_train[columns_for_minmax_scaling])
X_test[columns_for_minmax_scaling] = minmax.transform(X_test[columns_for_minmax_scaling])
```

```python
# Afficher la distribution des variables après mise à l'échelle
plt.figure(figsize=(12, 6))
plt.suptitle('Distribution des variables après mise à l\'échelle (MinMaxScaler)', fontsize=16)

for i, col in enumerate(columns_for_minmax_scaling, start=1):
    plt.subplot(2, 3, i)
    sns.histplot(X_train[col], bins=16)
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
# Sélectionner les colonnes dont les valeurs dépassent les bornes -1:1
out_of_bounds_columns = X_train.columns[(X_train.min() < -1) | (X_train.max() > 1)]

# Afficher les statistiques pour ces colonnes
for col in out_of_bounds_columns:
    print(f"Variable: {col}")
    print(f"Type: {X_train[col].dtype}")
    print(f"Min: {X_train[col].min()}")
    print(f"Max: {X_train[col].max()}")
    print(f"Variance: {X_train[col].var()}")
    print(f"Ecart type: {X_train[col].std()}")
    print(f"Moyenne: {X_train[col].mean()}")
    print("-" * 40)
```

## Machine Learning

### Simple Regression Linéaire

```python
lr_mae = LinearRegression()
lr_mae.fit(X_train, y_train)

y_pred = lr_mae.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R^2: {r2}")
```

### ElasticNet

```python
from sklearn.linear_model import ElasticNetCV

model_en = ElasticNetCV(l1_ratio=(0.1, 0.25, 0.5, 0.7, 0.75, 0.8, 0.85, 0.9, 0.99), 
                        alphas=(0.001, 0.01, 0.02, 0.025, 0.05, 0.1, 0.25, 0.5, 0.8, 1.0), 
                        cv = 8,
                        max_iter = 5000)
model_en.fit(X_train, y_train)
```

```python
# MSE
alphas = model_en.alphas_

plt.figure(figsize=(10, 10))

for i in range(model_en.mse_path_.shape[0]) :
    plt.plot(alphas, model_en.mse_path_[i,:,:].mean(axis=1),
             label='Moyenne pour l1_ratio= %.2f' %model_en.l1_ratio[i], linewidth=2)

plt.xlabel('Alpha')
plt.ylabel('Mean squared error')
plt.title('Mean squared error pour chaque $\lambda$')
plt.legend();
```

```python
# RMSE
pred_train = model_en.predict(X_train)
pred_test = model_en.predict(X_test)
print(np.sqrt(mean_squared_error(y_train, pred_train)))
print(np.sqrt(mean_squared_error(y_test, pred_test)))
```

```python
# Score
print('score train :', model_en.score(X_train, y_train))
print('score test :', model_en.score(X_test, y_test))
```

### XGBoost

```python
# Create the XGBoost regression model
model_xgb = xgb.XGBRegressor(
    n_estimators=100,   # Number of trees
    learning_rate=0.1,  # Step size shrinkage
    max_depth=6,        # Maximum depth of a tree
    subsample=0.8,      # Subsample ratio of the training instances
    colsample_bytree=0.8,  # Subsample ratio of columns when constructing each tree
    reg_alpha =0.1,          # L1 regularization term on weights
    reg_lambda =1.0         # L2 regularization term on weights
)

# Fit the model on the training data
model_xgb.fit(X_train, y_train)
```

```python
y_pred = model_xgb.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

fig, ax = plt.subplots(figsize=(10, 15)) 
xgb.plot_importance(model_xgb, ax=ax)
plt.show()
```

### GridSearch XGBRegressor

```python
param_grid = {
    'n_estimators': [100, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 6, 10],
    'subsample': [0.6, 1.0],
    'colsample_bytree': [0.6, 1.0],
    'alpha': [0, 0.1, 1.0],
    'lambda': [1.0, 2.0]
}

grid_search = GridSearchCV(estimator=xgb.XGBRegressor(), param_grid=param_grid, scoring='neg_mean_squared_error', cv=3, verbose=1, n_jobs=-1)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print(f"Best parameters found: {best_params}")

model_xgb = xgb.XGBRegressor(**best_params)
model_xgb.fit(X_train, y_train)
```

```python
scores = cross_val_score(model_xgb, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
rmse_scores = np.sqrt(-scores)
print(f"RMSE scores: {rmse_scores}")
print(f"Mean RMSE: {rmse_scores.mean()}, Standard Deviation RMSE: {rmse_scores.std()}")

fig, ax = plt.subplots(figsize=(10, 15)) 
xgb.plot_importance(model_xgb, ax=ax)
plt.show()
```
