```python
import pandas as pd
import numpy as np
```

```python
df_merge4 = pd.read_csv('data_merged_su.csv')
```

# Filtre

```python
columns_to_exclude = ['DeployedFromStation_Code', 'PlusCode_Code', 'DelayCodeId', 'Postcode_full', 'ProperCase', 'FRS', 'HourOfCall']
df_merge4_selection = df_merge4.drop(columns=columns_to_exclude)
                            
df_merge4_selection.head()
```

```python
# Trouver les valeurs NaN dans le DataFrame
nan_values = df_merge4_selection.isna()

# Trouver les valeurs null (None) dans le DataFrame
null_values = df_merge4_selection.isnull()

# Afficher les résultats
print("Valeurs NaN :")
display(nan_values)
print("\nValeurs Null :")
display(null_values)
```

Gestion des NaN de turnouttime seconds

```python
mean_TurnoutTimeSeconds = df_merge4_selection['TurnoutTimeSeconds'].mean()
df_merge4_selection['TurnoutTimeSeconds'] = df_merge4_selection['TurnoutTimeSeconds'].fillna(mean_TurnoutTimeSeconds)

df_merge4_1 = df_merge4_selection.dropna(subset=['CalYear'])
df_merge4_2 = df_merge4_1.dropna(subset=['TurnoutTimeSeconds'])
```

# Comparaison des TurnoutTimeSeconds par année (CalYear)

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

# Graphique à barres pour la moyenne des TurnoutTimeSeconds par année
plt.figure(figsize=(12, 10))
sns.barplot(x='CalYear', y='TurnoutTimeSeconds', data=df_merge4_2, estimator=np.mean)

# Ajouter les annotations de moyenne sur chaque barre
for index, row in df_merge4_2.groupby('CalYear')['TurnoutTimeSeconds'].mean().reset_index().iterrows():
    plt.text(row.name, row.TurnoutTimeSeconds + 0.5, round(row.TurnoutTimeSeconds, 2), color='black', ha="center")
    
    
plt.title('Moyenne des TurnoutTimeSeconds par année')
plt.xlabel('Année')
plt.ylabel('Temps de départ (secondes)')
plt.show()
```

```python
# Boîte à moustaches pour la distribution des TurnoutTimeSeconds par année
plt.figure(figsize=(12, 6))
sns.boxplot(x='CalYear', y='TurnoutTimeSeconds', data=df_merge4_2)
plt.title('Distribution des TurnoutTimeSeconds par année')
plt.xlabel('Année')
plt.ylabel('Temps de départ (secondes)')
plt.show()
```

# Comparaison des AttendanceTimeSeconds par année (CalYear)

```python
# Graphique à barres pour la moyenne des AttendanceTimeSeconds par année
plt.figure(figsize=(12, 6))
sns.barplot(x='CalYear', y='AttendanceTimeSeconds', data=df_merge4_2, estimator=np.mean)
plt.title('Moyenne des AttendanceTimeSeconds par année')
plt.xlabel('Année')
plt.ylabel('Temps de réponse (secondes)')
plt.show()


print("Les pompiers")
```

# Analyse des temps de réponse (AttendanceTimeSeconds) par station de déploiement (DeployedFromStation_Name)

```python
# Graphique à barres pour la moyenne des AttendanceTimeSeconds par station
plt.figure(figsize=(14, 8))
sns.barplot(x='DeployedFromStation_Name', y='AttendanceTimeSeconds', data=df_merge4_2, estimator=np.mean)
plt.title('Moyenne des AttendanceTimeSeconds par station')
plt.xlabel('Nom de la station')
plt.ylabel('Temps de réponse (secondes)')
plt.xticks(rotation=90)
plt.show()
```
