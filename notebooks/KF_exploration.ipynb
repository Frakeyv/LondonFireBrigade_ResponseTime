### Exploration variables de temps

On essaie de comparer les différentes valeurs de temps présentes.

```python
df = df.dropna(subset=["DateAndTimeMobile"])
```

```python
df['DateAndTimeCalled']= df['DateOfCall']+ [' '] + df['TimeOfCall']

df['DateAndTimeCalled']= pd.to_datetime(df['DateAndTimeCalled'])
df['DateAndTimeMobilised'] = pd.to_datetime(df['DateAndTimeMobilised'])
df['DateAndTimeMobile'] = pd.to_datetime(df['DateAndTimeMobile'])
df['DateAndTimeArrived'] = pd.to_datetime(df['DateAndTimeArrived'])
```

-------------

Compréhension de 'DateAndTimeCalled'

```python
dif_1 = (df['DateAndTimeMobilised'] - df['DateAndTimeCalled']) 
print('somme de la diff ', dif_1.sum()) # pourquoi est-ce négatif ???
print('moy de la diff ' , dif_1.mean())
print('std de la diff ' , dif_1.std())

test = (dif_1 > pd.Timedelta(0)).value_counts()
print(test)
```

```python
dif_2 = (df['DateAndTimeMobile']- df['DateAndTimeCalled'])
print('somme de la diff ', dif_2.sum()) # pourquoi est-ce négatif ???
print('moy de la diff ' , dif_2.mean())
print('std de la diff ' , dif_2.std())

test = (dif_2 > pd.Timedelta(0)).value_counts()
print(test)
```

```python
dif_3 = (df['DateAndTimeArrived']-df['DateAndTimeCalled']) 
print('somme de la diff ', dif_3.sum()) # pourquoi est-ce négatif ???
print('moy de la diff ' , dif_3.mean())
print('std de la diff ' , dif_3.std())

test = (dif_3 > pd.Timedelta(0)).value_counts()
print(test)
```

-------------------

Compréhension de DateAndtime Mobilised/Mobile/arrived

```python
dif_4 = (df['DateAndTimeArrived'] - df['DateAndTimeMobilised'])
print('somme de la diff ', dif_4.sum())
print('moy de la diff ' , dif_4.mean())
print('std de la diff ' , dif_4.std())

test = (dif_4 > pd.Timedelta(0)).value_counts()
print(test)
```

```python
dif_5 = (df['DateAndTimeMobile'] - df['DateAndTimeMobilised'])
print('somme de la diff ', dif_5.sum())
print('moy de la diff ' , dif_5.mean())
print('std de la diff ' , dif_5.std())

test = (dif_5 > pd.Timedelta(0)).value_counts()
print(test)
test = (dif_5.abs().dt.total_seconds().astype(int) == df.TurnoutTimeSeconds).value_counts()
print(test)
```

```python
dif_6 = (df['DateAndTimeArrived'] - df['DateAndTimeMobile'])
print('somme de la diff ', dif_6.sum())
print('moy de la diff ' , dif_6.mean())
print('std de la diff ' , dif_6.std())

test = (dif_6 > pd.Timedelta(0)).value_counts()
print(test)
test = (dif_6.abs().dt.total_seconds().astype(int) == df.TravelTimeSeconds).value_counts()
print(test)
```
